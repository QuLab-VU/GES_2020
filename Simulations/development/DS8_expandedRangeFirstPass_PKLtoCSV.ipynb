```python
DS8_ER1 = pd.concat([DS8_rest_ER1.reset_index(drop=True), 
                     DS8_dipdf_ER1.reset_index(drop=True), 
                     DS8_divdf_ER1.reset_index(drop=True),
                     DS8_dthdf_ER1], axis = 1)

DS8_ER2 = pd.concat([DS8_rest_ER2.reset_index(drop=True), 
                     DS8_dipdf_ER2.reset_index(drop=True), 
                     DS8_divdf_ER2.reset_index(drop=True),
                     DS8_dthdf_ER2], axis = 1)

DS8_ER3 = pd.concat([DS8_rest_ER3.reset_index(drop=True), 
                     DS8_dipdf_ER3.reset_index(drop=True), 
                     DS8_divdf_ER3.reset_index(drop=True),
                     DS8_dthdf_ER3], axis = 1)

DS8_ER4 = pd.concat([DS8_rest_ER4.reset_index(drop=True), 
                     DS8_dipdf_ER4.reset_index(drop=True), 
                     DS8_divdf_ER4.reset_index(drop=True),
                     DS8_dthdf_ER4], axis = 1)

DS8_ER5 = pd.concat([DS8_rest_ER5.reset_index(drop=True), 
                     DS8_dipdf_ER5.reset_index(drop=True), 
                     DS8_divdf_ER5.reset_index(drop=True),
                     DS8_dthdf_ER5], axis = 1)

DS8_ER6 = pd.concat([DS8_rest_ER6.reset_index(drop=True), 
                     DS8_dipdf_ER6.reset_index(drop=True), 
                     DS8_divdf_ER6.reset_index(drop=True),
                     DS8_dthdf_ER6], axis = 1)

DS8_ER7 = pd.concat([DS8_rest_ER7.reset_index(drop=True), 
                     DS8_dipdf_ER7.reset_index(drop=True), 
                     DS8_divdf_ER7.reset_index(drop=True),
                     DS8_dthdf_ER7], axis = 1)

DS8_ER8 = pd.concat([DS8_rest_ER8.reset_index(drop=True), 
                     DS8_dipdf_ER8.reset_index(drop=True), 
                     DS8_divdf_ER8.reset_index(drop=True),
                     DS8_dthdf_ER8], axis = 1)
```

```python
DS8_all_ER = pd.concat([DS8_ER1, DS8_ER2, DS8_ER3, DS8_ER4,
                      DS8_ER5, DS8_ER6, DS8_ER7, DS8_ER8], ignore_index = True)
DS8 = DS8_all_ER
```

```python
## Annotate with modified identifier (ease of plotting)
DS8['cell.line'] = np.where(DS8['p-value']>0.1, 'PC9-DS8', 'not.assigned')
DS8['param.pair'] = range(DS8.shape[0])

```

```python
## Create subsets of data frame (see below)
DS8_sig1 = DS8[['p-value', 'cell.line', 'DIP1', 'div1', 'dth1', 'param.pair']]
DS8_sig2 = DS8[['p-value', 'cell.line', 'DIP2', 'div2', 'dth2', 'param.pair']]

```

```python
DS8_sig1.head()
```

```python
## Rename columns of data frame subsets
DS8_sig1.rename(columns={'DIP1': 'DIP Rate',
                         'div1': 'Division Rate',
                         'dth1': 'Death Rate'},
                 inplace=True)
DS8_sig2.rename(columns={'DIP2': 'DIP Rate',
                         'div2': 'Division Rate',
                         'dth2': 'Death Rate'},
                inplace=True)

```

```python
## Separate subsets based on which DS8 state it resides in
DS8_sig1['Cell Line'] = np.where(DS8_sig1['cell.line'] == "PC9-DS8", 'PC9-DS8.1', 'not.assigned')
DS8_sig2['Cell Line'] = np.where(DS8_sig2['cell.line'] == "PC9-DS8", 'PC9-DS8.2', 'not.assigned')

## Concatenate datasets back together
DS8_sig_all = pd.concat([DS8_sig1, DS8_sig2])
```
