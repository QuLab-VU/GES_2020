```python
import numpy as np
import pandas as pd

# Read in pickled simulated datasets
df_DS6_expansion = pd.read_pickle('PC9-DS6_param-scan_Expansion.pkl')
```

```python
df_DS4_expansion['cell line'] = 'PC9.DS4'

df_DS4_expansion = df_DS4_expansion[['cell line', 'division rate', 'death rate', 'DIP rate', 
                                     'sim DIPs', 'KS p-value', 'AD p-value']]

KS_val = []
AD_val = []
for index, row in df_DS4_expansion.iterrows():
    KS_val.append(np.mean(np.array(df_DS4_expansion['KS p-value'][index])) + 2*np.std(np.array(df_DS4_expansion['KS p-value'][index])))
    AD_val.append(np.mean(np.array(df_DS4_expansion['AD p-value'][index])) + 2*np.std(np.array(df_DS4_expansion['AD p-value'][index])))

df_DS4_expansion['KS val'] = KS_val
df_DS4_expansion['AD val'] = AD_val

df_DS4_expansion = df_DS4_expansion[['cell line', 'division rate', 'death rate', 'DIP rate', 'KS val', 'AD val']]

df_DS4_expansion.head()
```

```python
# Function to label cell line if achieve a certain p-value threshold
def label_cell_line(df):
    if df['AD val'] > 0.05:
        return df['cell line']
    else:
        return "not.assigned"

# Add new column (corresponding to modified label - see above function)
df_DS4_expansion['cell line new'] = df_DS4_expansion.apply(lambda df_DS4_expansion: label_cell_line(df_DS4_expansion), axis = 1)

df_DS4_expansion.head()
```

```python
# Rename columns and save dataset
df_DS4_expansion.to_csv('DS4_expansionTest_tile.csv')
```
